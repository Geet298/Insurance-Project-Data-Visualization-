```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance.csv')
df.head(100)
```

```python
from matplotlib import pyplot as plt
```

```python
#1
#df.loc[: ,['bmi', 'charges']]
x = df['bmi']
y = df['charges']
color = df['smoker']
plt.scatter(x,y, c = color, alpha = 0.7) 
plt.title('BMI Vs Medical Expenses')
plt.xlabel('BMI of Patient')
plt.ylabel('Medical Expenses')
cbar = plt.colorbar(ticks=[0, 1])
cbar.set_ticklabels(['Nonsmoker', 'Smoker'])
plt.show()

```

```python
df[df.smoker == 0]
medical_cost = df[df.smoker == 0]['charges']
#print(medical_cost)
plt.hist(medical_cost, edgecolor='black')
num_bins = 12
plt.xlabel('Medical Expenses')
plt.ylabel('Number of Smoker')
plt.show()
```

```python
#2
num_bins = 12
df[df.smoker == 1]
smoker_medicalcost = df[df.smoker == 1]['charges']
plt.hist(smoker_medicalcost,num_bins, alpha = 0.6, edgecolor='black')
plt.show()

nonsmoker_medicalcost = df[df.smoker == 0]['charges']
plt.hist(nonsmoker_medicalcost, num_bins, alpha = 0.6, edgecolor='black')
plt.xlabel('Medical Expenses')
plt.ylabel('Number of Smoker and Nonsmoker Patient')
plt.legend(['Smoker', 'Nonsmoker'])
plt.title('Medical Expenses Distribution between Smokers and Non-smokers Patient');
```
